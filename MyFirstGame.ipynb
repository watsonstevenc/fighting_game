```python
#fighter definition
class Person:
    def __init__(self, name='Tom', age=20, power=10, health=100):
        self.first_name=name
        self.my_age=age
        self.power=power
        self.health=health
    
    #greeting function
    def greeting(self):
        print("Hi, my name is " + str(self.first_name) + " and I have " + str(self.health) + " health and " + str(self.power) + " power")
    
    #attack function
    def attack(self):
        return self.power
    
#basic game with two objects
def fight():
    #input data
    print('Player 1 input')
    p1=fight_inputs()
    print('Player 2 input')
    p2=fight_inputs()
    
    #check alive
    while p1.health>0 and p2.health>0:
        #p1 attack
        p2.health=p2.health-p1.attack()
        print(p1.first_name + ' attacks ' + p2.first_name + ' for ' + str(p1.power) + ' and drops their health to ' + str(p2.health))
        if p2.health<=0:
            return p1.first_name + ' wins!'
        #p2 attack
        p1.health=p1.health-p2.attack()
        print(p2.first_name + ' attacks ' + p1.first_name + ' for ' + str(p2.power) + ' and drops their health to ' + str(p1.health))
        if p1.health<=0:
            return p2.first_name + ' wins!'

#player inputs
def fight_inputs():
    p1=Person()
    print('Hello. What is your name?')
    p1.first_name=input()
    print('What is your health?')
    p1.health=int(input())
    print('What is your power?')
    p1.power=int(input())        
    p1.greeting()
    return p1

#run the game
fight()
```
